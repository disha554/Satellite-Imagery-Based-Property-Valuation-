```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader

import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm
```

```python
image_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

cnn_model = models.resnet18(pretrained=True)
cnn_model.fc = torch.nn.Identity()   # remove final classification layer
cnn_model = cnn_model.to(device)
cnn_model.eval()
```

```python
class SatelliteImageDataset(Dataset):
    def __init__(self, image_paths, transform):
        self.image_paths = image_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")
        img = self.transform(img)
        return img
```

```python
image_dataset = SatelliteImageDataset(
    train_df['image_path'].values,
    image_transform
)

image_loader = DataLoader(
    image_dataset,
    batch_size=32,
    shuffle=False,
    num_workers=2
)
```

```python
image_embeddings = []

with torch.no_grad():
    for batch in tqdm(image_loader, desc="Extracting CNN features"):
        batch = batch.to(device)
        features = cnn_model(batch)          # (B, 512)
        image_embeddings.append(features.cpu().numpy())

image_embeddings = np.vstack(image_embeddings)
```

```python
image_embeddings.shape
```

```python
train_df['cnn_features'] = list(image_embeddings)
```

### Image Visualisation


```python
def compute_green_cover(img_path):
    img = Image.open(img_path).convert("RGB").resize((224, 224))
    arr = np.array(img).astype(float)

    green_channel = arr[:, :, 1]
    total_intensity = arr.mean(axis=2)

    green_ratio = green_channel.mean() / (total_intensity.mean() + 1e-6)
    return green_ratio
```

```python
train_df['green_cover'] = [
    compute_green_cover(p) for p in tqdm(train_df['image_path'], desc="Computing green cover")
]
```

```python

plt.figure(figsize=(7,5))
plt.scatter(train_df['green_cover'], train_df['price'], alpha=0.4, s=8)
plt.xlabel("Green Cover Ratio")
plt.ylabel("Price")
plt.title("Effect of Green Cover on Property Price")
plt.show()
```

Properties surrounded by higher vegetation density exhibit consistently higher prices, confirming the positive valuation effect of greenery.

```python
def compute_urban_density(img_path):
    img = Image.open(img_path).convert("L").resize((224, 224))
    arr = np.array(img).astype(float)
    return arr.var()
```

```python
train_df['urban_density'] = [
    compute_urban_density(p) for p in tqdm(train_df['image_path'], desc="Computing urban density")
]
```

```python
plt.figure(figsize=(7,5))
plt.scatter(train_df['urban_density'], train_df['price'], alpha=0.4, s=8)
plt.xlabel("Urban Density (Texture Variance)")
plt.ylabel("Price")
plt.title("Effect of Urban Density on Property Price")
plt.show()
```

High urban density, reflecting congestion and visual clutter, is negatively associated with property prices.

```python
def compute_water_score(img_path):
    img = Image.open(img_path).convert("RGB").resize((224, 224))
    arr = np.array(img).astype(float)

    blue = arr[:, :, 2]
    red = arr[:, :, 0]
    green = arr[:, :, 1]

    # Blue dominance heuristic
    water_score = blue.mean() / (red.mean() + green.mean() + 1e-6)
    return water_score
```

```python
train_df['water_score'] = [
    compute_water_score(p) for p in tqdm(train_df['image_path'], desc="Computing water score")
]
```

```python
plt.figure(figsize=(7,5))
plt.scatter(train_df['water_score'], train_df['price'], alpha=0.4, s=8)
plt.xlabel("Water Presence Score")
plt.ylabel("Price")
plt.title("Effect of Water Proximity on Property Price")
plt.show()
```

Properties with visual proximity to water bodies show clear price premiums, even beyond the binary waterfront indicator.

### Grad Cam

```python
import torch
import torch.nn.functional as F
from torchvision import models, transforms
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

cnn = models.resnet18(pretrained=True)
cnn.fc = torch.nn.Identity()   # IMPORTANT
cnn = cnn.to(device)
cnn.eval()
```

```python
gradients = None
activations = None

def backward_hook(module, grad_input, grad_output):
    global gradients
    gradients = grad_output[0]

def forward_hook(module, input, output):
    global activations
    activations = output
```

```python
target_layer = cnn.layer4[-1].conv2
target_layer.register_forward_hook(forward_hook)
target_layer.register_backward_hook(backward_hook)
```

```python
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
def load_image(image_path):
    img = Image.open(image_path).convert("RGB")
    img_tensor = transform(img).unsqueeze(0).to(device)
    return img_tensor
```

```python
def generate_gradcam(img_tensor):
    global gradients, activations

    output = cnn(img_tensor)

    # Proxy objective for regression-style usage
    score = output.mean()

    cnn.zero_grad()
    score.backward()

    pooled_grads = torch.mean(gradients, dim=[0, 2, 3])

    for i in range(activations.shape[1]):
        activations[:, i, :, :] *= pooled_grads[i]

    heatmap = torch.mean(activations, dim=1).squeeze()
    heatmap = F.relu(heatmap)
    heatmap /= torch.max(heatmap)

    return heatmap.cpu().detach().numpy()
```

```python
def overlay_gradcam(image_path, heatmap):
    image = cv2.imread(image_path)
    image = cv2.resize(image, (224,224))

    heatmap = cv2.resize(heatmap, (224,224))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    overlay = cv2.addWeighted(image, 0.6, heatmap, 0.4, 0)
    return overlay
```

```python
# Example: first valid image
sample_img_path = train_df.iloc[0]["image_path"]

img_tensor = load_image(sample_img_path)
heatmap = generate_gradcam(img_tensor)
overlay = overlay_gradcam(sample_img_path, heatmap)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Grad-CAM Visualization")
plt.show()
```

```python
cv2.imwrite(
    "/content/drive/MyDrive/gradcam_sample.png",
    overlay
)
```

Grad-CAM was used to provide visual interpretability for the CNN-based satellite image features by highlighting regions of an image that most influenced the learned representations. The heatmaps showed strong activation over meaningful areas such as green spaces, water bodies, and dense urban regions, confirming that the model focuses on relevant environmental features when contributing to property price prediction.
